## Plot the data points

NORMAL VISUALIZATION OF DATA

```python
# converting attributes into list
x =data['Original Price (Rs.)'].tolist()
y =data['Discounted Price (Rs.)'].tolist()
n=np.size(y)
```

```python
#correlation code between orignal price and discounted price
df=pd.DataFrame(data['Original Price (Rs.)'],data['Discounted Price (Rs.)'])
corr=df.corr()
print(corr)
```

```python
# Normalize input data to handle the outlier
x_mean = np.mean(x)
x_std = np.std(x)
x_normalized = (x - x_mean) / x_std
print(x_mean)
print(x_std)
print(x_normalized)
```

## CALCULATING AND UPATING THE INTERCEPT AND SLOPE

```python
#initialize the parameters
a0 = 0                  #intercept
a1 = 1                  #Slope
lr = 0.0001             #Learning rate
iterations = 1000       # Number of iterations
error = []              # Error array to calculate cost for each iterations.
for itr in range(iterations):
    error_cost = 0
    cost_a0 = 0
    cost_a1 = 0
    for i in range(len(x_normalized)):
        #linear equation predicted value
        y_pred = a0+a1*x_normalized[i]               # predict value for given x
        #cost function caluculation
        error_cost = error_cost +(y[i]-y_pred)**2
        for j in range(len(x_normalized)):
            #gradient descent calculation
            partial_wrt_a0 = -2 *(y[j] - (a0 + a1*x_normalized[j]))    #partial derivative w.r.t a0
            partial_wrt_a1 = (-2*x_normalized[j])*(y[j]-(a0 + a1*x_normalized[j]))           #partial derivative w.r.t a1
            cost_a0 = cost_a0 + partial_wrt_a0      #calculate cost for each number and add
            cost_a1 = cost_a1 + partial_wrt_a1      #calculate cost for each number and add
        a0 = a0 - lr * cost_a0    #update a0
        a1 = a1 - lr * cost_a1    #update a1
        print(itr,a0,a1)          #Check iteration and updated a0 and a1
    error.append(error_cost)      #Append the data in array
```

PRINTING UPDATED VALUES OF INTERCEPT AND SLOP

```python
#printing out comes to intercept and slope value
print(a0)
print(a1)
```

Predicting the values

```python
pred = a0 + a1 * x_normalized
print(pred)
```

```python
plt.scatter(x_normalized,y,color = 'red')
plt.plot(x_normalized,pred, color = 'green')
plt.xlabel("ORIGINAL PRICE")
plt.ylabel("DISCOUNTED PRICE")
```

Analyze the performance of the model by calculating the mean squared error

```python
error1 = y - pred
se = np.sum(error1 ** 2)
mse = se/n
print("mean squared error is", mse)
```
